#### Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
url_hackathon ='https://mlh.io/seasons/na-2020/events'
````

````python
#your code

response_hack = requests.get(url_hackathon)
content_hack = response_hack.content
print(response_hack)
````

````python
soup_hack = BeautifulSoup(content_hack, 'lxml')
````

````python
select_name = soup_hack.select('h3[class=event-name]')
event_name = [element.text for element in select_name]
event_name
````

````python
select_date = soup_hack.select('p[class=event-date]')
event_date = [element.text.strip() for element in select_date]
event_date
````

````python
select_location = soup_hack.select('div[class=event-location]')
event_location = [element.text.strip(' \n').split('\n') for element in select_location]
event_location
````

````python
event_location[0] # no puedo borrar espacios intermedios
````

````python
len(event_name), len(event_date), len(event_location)
````

````python
df_hack = pd.DataFrame(zip(event_name, event_date, event_location),
             columns=['Title', 'Date', 'City, Country'])

df_hack = df_hack[0:25]
df_hack
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
soup_wiki= BeautifulSoup(content_wiki, 'lxml')
````

````python
languages_all = soup_wiki.select('div[class=central-featured] > div.central-featured-lang > a ')
languages_all
````

````python
pd.DataFrame(languages)
````
